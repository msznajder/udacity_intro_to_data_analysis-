```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, ls=":", color="salmon")
axes.plot(x, y*3, marker=",")
axes.set_xlim([0, 100])
axes.set_ylim(0, 100)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([0, 5])
plt.subplot(1, 2, 2)
plt.plot(y, x)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y, markersize=2, marker="o", lw=3)
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].set_title("title")
axes[1].plot(y, x)
fig.tight_layout()
```

```python
plt.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize=(12, 8))
axes.plot(x, y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, marker="*")
```

```python
plt.scatter(x, x)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 0.5, 0.5])
axes2 = fig.add_axes([0.5, 0.5, 0.5, 0.5])
axes1.plot(x, y, label="xxx", ls="--")
axes2.plot(y, x)
axes1.legend(loc=0)
fig.savefig("test.png")
```
